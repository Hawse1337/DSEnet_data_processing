```python
import numpy as np
import scipy.io.wavfile as wf
import os
import matplotlib.pyplot as plt
```

```python
file_names = os.listdir('./clean_testset_wav_16k')
f2 = os.listdir('./noisy_testset_wav_16k')

if len(f2) != len(file_names):
    print('different')
```

```python
for f_name in file_names:
    SR,clean_wav = wf.read('./clean_testset_wav_16k/%s'%(f_name))
    SR,noisy_wav = wf.read('./noisy_testset_wav_16k/%s'%(f_name))
    if SR != 16000:
        print(f_name)
    wf.write('./test_noise_set_16k/%s'%(f_name),SR,noisy_wav-clean_wav)
```

```python
plt.plot(clean_wav)
plt.show()
plt.plot(noisy_wav)
plt.show()
plt.plot(noisy_wav-clean_wav)
plt.show()
```
