```python
def time_shift(indata,dt,delay):
    n=int(delay/dt); dtt=delay-n*dt
    buffer_size = len(indata) - abs(n)
    # dt-deltaT 만큼 옮길때(T)
    od=np.array([]);sp=np.array([]);spp=np.array([])
    if dtt>=0:
        print('dtt>=0',n)
        if n == 0:
            od=np.array(indata)
            sp=np.array(indata)
        else :
            od=np.array(indata[n:])
            sp=np.array(indata[:-n])            
        
        mp = (sp[1:] - sp[:-1] )/ dt
#         spp = mp*dtt+sp[1:]
        spp=np.append(sp[0],mp*dtt+sp[1:])
#         spp=np.append(mp*dtt+sp[:-1],sp[-1])
    else:
        print('dtt<0')
        if n == 0:
            od=np.array(indata)
            sp=np.array(indata)
        else :
            od=np.array(indata[0:-abs(n)])
            sp=np.array(indata[abs(n):])#-1-1:-1])
            #print(len(od),'####',len(sp))
        
        
        mp = (sp[:-1] - sp[1:]) / dt
        spp=np.append(mp*dtt+sp[:-1],sp[-1])
#         spp = mp*dtt+sp[:-1]
    return od,spp

def delay_time(dd):
    dt=[]
    d = dd*0.01 #m/cm
    v = 340 #m/s
    rad = np.pi/180
    for i in range(36):
        delta=(d*np.cos(i*10*rad))/v
        dt.append(delta)
        
    return dt

def norm(inp):
    M = np.amax(np.absolute(inp))
    return inp/M

def qnorm(inp):
    M = np.amax(np.absolute(inp))
    return inp*(32767/M).astype(np.int16)
#     if M >32767 :
#         return inp*(32767/M).astype(np.int16)
#     else:
#         return inp.astype(np.int16)
```
